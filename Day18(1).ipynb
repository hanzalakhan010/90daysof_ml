- compard pca results with tsne using iris dataset


```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
df = load_iris(as_frame=True)
df = df.frame
y = df.target
df = df.drop(columns=['target'])
```

```python
#Scaling the data
standard_scaler = StandardScaler()
x_std_scaled = standard_scaler.fit_transform(df)
mm_scaler = MinMaxScaler()
x_mm_scaled = mm_scaler.fit_transform(df)
```

```python
#Applying PCA on standart scaled data
pca = PCA(n_components=2)
xy_std  = pca.fit_transform(x_std_scaled)
plt.scatter(xy_std[:,0],xy_std[:,1],c =y ,cmap='Accent')
plt.show()
```

```python
pca = PCA(n_components=2)
xy_mm  = pca.fit_transform(x_mm_scaled)
plt.scatter(xy_mm[:,0],xy_mm[:,1],c =y ,cmap='Accent')
plt.show()
```

```python
x_tsne = TSNE(n_components=2)
xy_std = x_tsne.fit_transform(x_std_scaled)
plt.scatter(xy_std[:,0],xy_std[:,1],c =y ,cmap='Accent')
plt.show()
```

```python
x_tsne = TSNE(n_components=2)
xy_mm = x_tsne.fit_transform(x_mm_scaled)
plt.scatter(xy_mm[:,0],xy_mm[:,1],c =y ,cmap='Accent')
plt.show()
```
