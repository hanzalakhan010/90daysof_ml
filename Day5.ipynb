```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./datasets/iris/Iris.csv')
df.head()
# df.shape
```

## Guide Line
**Tasks for Day 5**
1. Understand the Basics of Data Exploration
Why Explore the Data?
    To identify patterns, trends, or anomalies.
    To understand the distribution of data.
    To identify relationships or correlations between features.

Key Questions to Ask:
    What are the central tendencies and spreads (mean, median, variance)?
    Are there any outliers or anomalies?
    Are features correlated with one another?
    Are there patterns across categorical or time-based features?

```python
#What are the central tendencies and spreads (mean, median, variance)?
df.groupby('Species').agg({
    'SepalLengthCm':['min','max','std','var'],
    'SepalWidthCm':['min','max','std'],
    'PetalLengthCm':['min','max','std'],
    'PetalWidthCm':['min','max','std'],
}).transpose()

# df.describe()
```

```python
#Are there any outliers or anomalies?
#Box plot
sns.boxplot(x = df['PetalLengthCm']) # right skewed 
sns.boxplot(x = df['SepalLengthCm']) # normally distributed 
sns.boxplot(x = df['PetalLengthCm']) # right skewed 
sns.boxplot(x = df['PetalLengthCm']) # right skewed 
```

```python
#Are features correlated with one another?
# scatter plot
#comparing petal_length with petal_width
plt.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
#positive corelation
```

```python
#Comparing sepal_length with sepal_width
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'])
#no corelation
```

```python
#Comparing petal_length with sepal_length
plt.scatter(df['PetalLengthCm'],df['SepalLengthCm'])
#strong positive corelation
```

```python
# comparing petalwith with sepal width
plt.scatter(df['PetalWidthCm'],df['SepalWidthCm'])
#weak positive corelation
```

```python
#using heatmap to visualise correlation between numerial values 
numerical = df.drop(columns = ['Species']).corr()
sns.heatmap(numerical, annot=True, cmap='coolwarm')
```
