# **Agenda**
  - PCA(prinipal component alnalysis)   
  - t-SNE(t-Distributed Stochastic Neighbor Embedding)
  - UMAP(Uniform Manifold Approximation and Projection)


### **Comparison Table**

| Aspect               | PCA                     | t-SNE                     | UMAP                      |
|----------------------|-------------------------|---------------------------|---------------------------|
| **Method**           | Linear                 | Non-linear                | Non-linear                |
| **Goal**             | Maximize variance      | Preserve local structure  | Preserve local + global structure |
| **Speed**            | Fast                   | Slow                      | Faster than t-SNE         |
| **Scalability**      | Scales well            | Limited to ~10,000 samples | Handles millions of samples |
| **Output Dimensions**| Any                    | Typically 2 or 3          | Typically 2 or 3          |
| **Hyperparameters**  | Minimal                | Sensitive (e.g., perplexity) | Intuitive (e.g., `n_neighbors`, `min_dist`) |
| **Global Structure** | Good                   | Poor                      | Better than t-SNE         |
| **Local Structure**  | Weak                   | Excellent                 | Excellent                 |
| **Interpretability** | High                   | Low                       | Moderate                  |

---

### **When to Use Each**:
- **PCA**:
  - Data preprocessing or dimensionality reduction for ML pipelines.
  - When computational speed is a priority.
  - When you need interpretable components (e.g., eigenvectors).

- **t-SNE**:
  - Visualizing small datasets or embeddings where clustering is important.
  - Highlighting local groupings or separations.

- **UMAP**:
  - Exploring large, complex datasets with a mix of global and local relationships.
  - Preprocessing for clustering or visualization tasks.
  - Faster, large-scale alternatives to t-SNE.

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
df = load_breast_cancer(as_frame=True)
df = df.frame
y = df['target']
df.drop(columns=['target'])
```

```python
#scaling the data
scaler = StandardScaler()
df = scaler.fit_transform(df)
```

```python
#performing PCA
pca = PCA(n_components = 2)
xy = pca.fit_transform(df)
```

```python
# plotting the scatter plot
plt.scatter(xy[:,0],xy[:,1],c = y,cmap='Accent')
plt.show()
```

```python
#Appliying tsne
X_tsne = TSNE(n_components=2,random_state=3,perplexity=30,n_iter = 10000)
xy = X_tsne.fit_transform(df)
```

```python
plt.scatter(xy[:,0],xy[:,1],c = y, cmap='Accent')
```
